```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
```

```python
# Constructing the LL;
a = Node('A')
b = Node('B')
c = Node('C')
d = Node('D')

a.next = b
b.next = c
c.next = d
```

### Question-1: Traversing a Linkedin List (Iteractive and Recursive):

```python
def print_list_iterative(head):
    current = head 
    string = ""
    while current is not None:
        if current.next is not None:
            extra = " --> "  
        else:
            extra = " --> None "
        string += str(current.val) + extra
        #print(current.val)
        current = current.next        
    
    print(string)
print_list_iterative(a)    

print(" ")

def print_list_recursive(head, string = ""):
    if head==None:
        print(string)
        return 
    
    if head.next is None:
        extra = " --> None "
    else:
        extra = " --> "  
    string += str(head.val) + extra
    print_list_recursive(head.next, string)
    
print_list_recursive(a)
```

### Question-2: Write a function, that takes in the head of a linked list as an argument. The function should return a list containing all values of the nodes in the linked list:

```python
def linked_list_values_iterative(head):
    answer = [] 
    current = head
    while current is not None:
        answer.append(current.val)
        current = current.next
    return answer

a = Node("a")
b = Node("b")
c = Node("c")
d = Node("d")

a.next = b
b.next = c
c.next = d
print(linked_list_values_iterative(a)) 


x = Node("x")
y = Node("y")
x.next = y
print(linked_list_values_iterative(x))

q = Node("q")
# q
print(linked_list_values_iterative(q)) 

print(linked_list_values_iterative(None))



print(" ")



def fill_values(head, answer):
    if head is None:
        return 
    
    answer.append(head.val)
    fill_values(head.next, answer)
    
def linked_list_values_recursive(head):
    answer = []
    fill_values(head, answer)
    return answer


print(linked_list_values_recursive(a)) 
print(linked_list_values_recursive(x))
print(linked_list_values_iterative(q)) 
print(linked_list_values_iterative(None))
```

### Question-3: Write a function, sum_list, that takes in the head of a linked list containing numbers as an argument. The function should return the total sum of all values in the linked list.

```python
def sum_list_iterative(head):
    current = head 
    currSum = 0
    while current!=None:
        currSum += current.val
        current = current.next
    return currSum    
        
a = Node(2)
b = Node(8)
c = Node(3)
d = Node(-1)
e = Node(7)

a.next = b
b.next = c
c.next = d
d.next = e
# 2 -> 8 -> 3 -> -1 -> 7
print(sum_list_iterative(a)) # 19


x = Node(38)
y = Node(4)

x.next = y
# 38 -> 4
print(sum_list_iterative(x)) # 42

z = Node(100)
# 100
print(sum_list_iterative(z)) # 100

print(sum_list_iterative(None)) # 0



print(" ")



def sum_list_recursive(head):
    if head is None:
        return 0
    
    return head.val + sum_list_recursive(head.next)


print(sum_list_recursive(a)) 
print(sum_list_recursive(x))
print(sum_list_recursive(z)) 
print(sum_list_recursive(None))
```

### Question-4: Write a function, linked_list_find, that takes in the head of a linked list and a target value. The function should return a boolean indicating whether or not the linked list contains the target.

```python
def linked_list_find(head, target):
    current = head 
    while current!=None:
        if current.val==target:
            return True
        current = current.next

    return False 


a = Node("a")
b = Node("b")
c = Node("c")
d = Node("d")
a.next = b
b.next = c
c.next = d
# a -> b -> c -> d
print(linked_list_find(a, "c")) # True

a = Node("a")
b = Node("b")
c = Node("c")
d = Node("d")
a.next = b
b.next = c
c.next = d
# a -> b -> c -> d

print(linked_list_find(a, "d")) # True

a = Node("a")
b = Node("b")
c = Node("c")
d = Node("d")
a.next = b
b.next = c
c.next = d
# a -> b -> c -> d
print(linked_list_find(a, "q")) # False

node1 = Node("jason")
node2 = Node("leneli")
node1.next = node2
# jason -> leneli
print(linked_list_find(node1, "jason")) # True

node1 = Node(42)
# 42
print(linked_list_find(node1, 42)) # True

node1 = Node(42)
# 42
print(linked_list_find(node1, 100)) # False


print(" ")

def linked_list_find_recursive(head, target):
    if head is None:
        return False

    
    return head.val==target or linked_list_find_recursive(head.next, target)

a = Node("a")
b = Node("b")
c = Node("c")
d = Node("d")
a.next = b
b.next = c
c.next = d
# a -> b -> c -> d
print(linked_list_find_recursive(a, "c")) 

a = Node("a")
b = Node("b")
c = Node("c")
d = Node("d")
a.next = b
b.next = c
c.next = d
# a -> b -> c -> d
print(linked_list_find_recursive(a, "d"))

a = Node("a")
b = Node("b")
c = Node("c")
d = Node("d")
a.next = b
b.next = c
c.next = d
# a -> b -> c -> d
print(linked_list_find_recursive(a, "q")) 


node1 = Node("jason")
node2 = Node("leneli")
node1.next = node2
# jason -> leneli
print(linked_list_find_recursive(node1, "jason")) # True

node1 = Node(42)
# 42
print(linked_list_find_recursive(node1, 42)) # True

node1 = Node(42)
# 42
print(linked_list_find_recursive(node1, 100)) # False
```

### Question-5: Write a function, get_node_value, that takes in the head of a linked list and an index. The function should return the value of the linked list at the specified index. If there is no node at the given index, then return None.

```python
def get_node_value_iterative(head, targetIndex):
    currIndex = 0 
    current = head
    while current!=None:
        if currIndex == targetIndex:
            return current.val
        current = current.next
        currIndex += 1
    return None    

a = Node("a")
b = Node("b")
c = Node("c")
d = Node("d")
a.next = b
b.next = c
c.next = d
# a -> b -> c -> d
print(get_node_value_iterative(a, 2)) # 'c'


a = Node("a")
b = Node("b")
c = Node("c")
d = Node("d")
a.next = b
b.next = c
c.next = d
# a -> b -> c -> d
print(get_node_value_iterative(a, 3)) # 'd'


a = Node("a")
b = Node("b")
c = Node("c")
d = Node("d")
a.next = b
b.next = c
c.next = d
# a -> b -> c -> d
print(get_node_value_iterative(a, 7)) # None

node1 = Node("banana")
node2 = Node("mango")
node1.next = node2
# banana -> mango
print(get_node_value_iterative(node1, 0)) # 'banana'

node1 = Node("banana")
node2 = Node("mango")
node1.next = node2
# banana -> mango
print(get_node_value_iterative(node1, 1)) # 'mango'


print(" ")


def get_node_value_recursive(head, targetIndex):
    if head is None:
        return None

    if targetIndex==0:
        return head.val

    return get_node_value_recursive(head.next, targetIndex-1)


a = Node("a")
b = Node("b")
c = Node("c")
d = Node("d")
a.next = b
b.next = c
c.next = d
# a -> b -> c -> d
print(get_node_value_recursive(a, 2)) # 'c'


a = Node("a")
b = Node("b")
c = Node("c")
d = Node("d")
a.next = b
b.next = c
c.next = d
# a -> b -> c -> d
print(get_node_value_recursive(a, 3)) # 'd'


a = Node("a")
b = Node("b")
c = Node("c")
d = Node("d")
a.next = b
b.next = c
c.next = d
# a -> b -> c -> d
print(get_node_value_recursive(a, 7)) # None

node1 = Node("banana")
node2 = Node("mango")
node1.next = node2
# banana -> mango
print(get_node_value_recursive(node1, 0)) # 'banana'

node1 = Node("banana")
node2 = Node("mango")
node1.next = node2
# banana -> mango
print(get_node_value_recursive(node1, 1)) # 'mango'
```

### Question-6: Write a function, reverse_list, that takes in the head of a linked list as an argument. The function should reverse the order of the nodes in the linked list in-place and return the new head of the reversed linked list.

```python
def reverse_list(head):
    current = head 
    previous = None
    while current!=None:
        next = current.next
        current.next = previous 
        previous = current
        current = next 
    return previous

a = Node("a")
b = Node("b")
c = Node("c")
d = Node("d")
e = Node("e")
f = Node("f")
a.next = b
b.next = c
c.next = d
d.next = e
e.next = f
# a -> b -> c -> d -> e -> f
newHead = reverse_list(a) # f -> e -> d -> c -> b -> a
print(print_list_iterative(newHead)) 

x = Node("x")
y = Node("y")
x.next = y
# x -> y
newHead = reverse_list(x) 
print(print_list_iterative(newHead)) 

p = Node("p")
# p
newHead = reverse_list(p) # p
print(print_list_iterative(newHead)) 


print(" ")


def reverse_list_recursive(head, previous = None):
    if head is None:
        return previous
    
    next = head.next 
    head.next = previous
    previous = head 
    head = next 
    return reverse_list_recursive(head, previous)


a = Node("a")
b = Node("b")
c = Node("c")
d = Node("d")
e = Node("e")
f = Node("f")
a.next = b
b.next = c
c.next = d
d.next = e
e.next = f
newHead = reverse_list_recursive(a) # f -> e -> d -> c -> b -> a
print(print_list_iterative(newHead)) 

x = Node("x")
y = Node("y")
x.next = y
# x -> y
newHead = reverse_list_recursive(x) 
print(print_list_iterative(newHead)) 


p = Node("p")
# p
newHead = reverse_list_recursive(p) # p
print(print_list_iterative(newHead)) 
```

### Question-7: Write a function, zipper_lists, that takes in the head of two linked lists as arguments. The function should zipper the two lists together into single linked list by alternating nodes. If one of the linked lists is longer than the other, the resulting list should terminate with the remaining nodes. The function should return the head of the zippered linked list. Do this in-place, by mutating the original Nodes. You may assume that both input lists are non-empty.

```python
def zipper_lists(head1, head2):
    
    # the tail pointer helps me construct the linkedin list
    tail = head1
    
    current1 = head1.next
    current2 = head2
    count = 0
    
    while current1!=None and current2!=None:
        if count%2==0:
            tail.next = current2
            current2 = current2.next 
        else:
            tail.next = current1
            current1 = current1.next 
        
        tail = tail.next 
        count += 1
    
    if current1 is None:
        tail.next = current2
    if current2 is None:
        tail.next = current1
        
    return head1   


a = Node("a")
b = Node("b")
c = Node("c")
a.next = b
b.next = c
# a -> b -> c
x = Node("x")
y = Node("y")
z = Node("z")
x.next = y
y.next = z
# x -> y -> z
newHead = zipper_lists(a, x)
print(print_list_iterative(newHead)) 
# a -> x -> b -> y -> c -> z


a = Node("a")
b = Node("b")
c = Node("c")
d = Node("d")
e = Node("e")
f = Node("f")
a.next = b
b.next = c
c.next = d
d.next = e
e.next = f
# a -> b -> c -> d -> e -> f

x = Node("x")
y = Node("y")
z = Node("z")
x.next = y
y.next = z
# x -> y -> z
newHead = zipper_lists(a, x)
print(print_list_iterative(newHead)) 
# a -> x -> b -> y -> c -> z -> d -> e -> f


s = Node("s")
t = Node("t")
s.next = t
# s -> t
one = Node(1)
two = Node(2)
three = Node(3)
four = Node(4)
one.next = two
two.next = three
three.next = four
# 1 -> 2 -> 3 -> 4
newHead = zipper_lists(s, one)
print(print_list_iterative(newHead)) 
# s -> 1 -> t -> 2 -> 3 -> 4


w = Node("w")
# w
one = Node(1)
two = Node(2)
three = Node(3)
one.next = two
two.next = three
# 1 -> 2 -> 3 
newHead = zipper_lists(w, one)
print(print_list_iterative(newHead)) 
# w -> 1 -> 2 -> 3


print(" ")

```
